```python
def euclidean_dist(x1,y1,x2,y2):

  dist =  ( ((x1 - x2)**2) + ((y1 - y2)**2) )** 0.5 
  return dist

def manhattan_dist(x1,y1,x2,y2):

  dist = abs(x1 - x2) + abs(y1 - y2)
  return dist

def chessboard_dist(x1,y1,x2,y2):

  dist = max(abs(x1 - x2),abs(y1 - y2))
  return dist

def minkowski_dist(x1,y1,x2,y2,p):
  
  dist =  ( (abs(x1 - x2)**p) + (abs(y1 - y2)**p) )**(1/p) 
  return dist

def chebyshev_dist(x1,y1,x2,y2):

  dist = max(abs(x1 - x2),abs(y1 - y2))
  return dist

def cosine_dist(x1,y1,x2,y2):
  
  AB = x1*x2 + y1*y2;
  norm_A = (x1*x1 + y1*y1)**0.5
  norm_B = (x2*x2 + y2*y2)**0.5

  dist = 1-(AB/(norm_A*norm_B))
  return dist

def correlation_dist(x1,y1,x2,y2):
    p1=[x1,y1]
    p2=[x2,y2]
    norm_p1 = 0
    norm_p2 = 0
  
    for i in range(len(p1)):
        norm_p1 += (p1[i] - st.mean(p1))**2
        norm_p2 += (p2[i] - st.mean(p2))**2
      
    norm_p1 = norm_p1**0.5
    norm_p2 = norm_p2**0.5
  
    s = 0
    for i in range(len(p1)):
      s += (p1[i] - st.mean(p1))*(p2[i] - st.mean(p2))
    distance = 1 - s/(norm_p1*norm_p2)
    return distance


def bray_curtis_distance(x1,y1,x2,y2):
   
    p1=[x1,y1]
    p2=[x2,y2]
    s1 = abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])
    s2 = abs(p1[0] + p2[0]) + abs(p1[1] + p2[1])
  
    dist = s1/s2
    return dist

def canberra_distance(x1,y1,x2,y2):
    p1=[x1,y1]
    p2=[x2,y2]
    dist = 0
    s1 = abs(p1[0] - p2[0])
    s2 = abs(p1[0] + p2[0])
    dist += s1/s2
    s1 = abs(p1[1] - p2[1])
    s2 = abs(p1[1] + p2[1])
    dist += s1/s2

    return dist
```

```python
def mer(true_cl,pred_cl):
    error = 0
    for i in range(len(true_cl)):
        if(true_cl[i] != pred_cl[i]):
            error += 1

    mer_val = error/len(true_cl)
    return mer_val


def me(true_cl,pred_cl,d):
    error=0
    error_dist=0
    for i in range(len(true_cl)):
        if(true_cl[i] != pred_cl[i]):
            error += 1
            error_dist+=d[i]
    mean_error_val=error_dist/error
    return mean_error_val
```

```python
# Provide 60% of instances 

X = iris.data[:, :2]  # we only take the first two features
y = iris.target 

# For Class 0
# taking instances from 0 to 29 for training
Class0_1 = X[0:30,0]
Class0_2 = X[0:30,1]

# For Class 1 
# taking instances from 50 to 79 for training
Class1_1 = X[50:80,0]
Class1_2 = X[50:80,1]

# For Class 2
# taking instances from 100 to 129 for training
Class2_1 = X[100:130,0]
Class2_2 = X[100:130,1]

# Finding centroid for each class
# Centroid of Class 0
sentosa_s_l = st.mean(Class0_1)
sentosa_s_w = st.mean(Class0_2)

# Centroid of Class 1
versicolor_s_l = st.mean(Class1_1)
versicolor_s_w = st.mean(Class1_2)

# Centroid of Class 2
virginica_s_l = st.mean(Class2_1)
virginica_s_w = st.mean(Class2_2)

centroid = np.array([[sentosa_s_l, sentosa_s_w],
                        [versicolor_s_l, versicolor_s_w],
                        [virginica_s_l, virginica_s_w]])

#print(Class2_1)
```

```python
# row slicing & column slicing

C_x = np.concatenate((X[30:50,0], X[80:100,0], X[130:150,0] ))
C_y = np.concatenate((X[30:50,1], X[80:100,1], X[130:150,1] ))
```

```python
# euclidean distance

import sys 
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae

mer_list=[]
mean_error_list=[]
mse_list=[]
mae_list=[]
    
min_dist = sys.maxsize
predicted_dist=[0]*len(C_x)

predicted_class = [0]*len(C_x)
true_class= np.concatenate((y[30:50], y[80:100], y[130:150] ))

for i in range (len(C_x)):
        for j in range(0,3):

            distance = euclidean_dist(centroid[j,0],centroid[j,1],
                                      C_x[i],C_y[i])    
            if(min_dist > distance):
                min_dist = distance
                predicted_dist[i] = distance 
                tag = j;
                
        predicted_class[i] = tag;
        min_dist = sys.maxsize    #reset min_dist
        
#print(predicted_class)    
        
#True Class vs Predicted Class display
x={'True_class': true_class, 'Predicted_class': predicted_class}
df=pd.DataFrame(x)
print(df)     

actual_dist = [0]*len(C_x);

for i in range (len(C_x)):
          actual_dist[i] = euclidean_dist(centroid[int(true_class[i]),0],centroid[int(true_class[i]),1],
                                               C_x[i],C_y[i])

# Accuracy Calculations
mer_error = mer(true_class,predicted_class)
mean_error=me(true_class,predicted_class,predicted_dist)
mse_error = mse(actual_dist,predicted_dist)
mae_error = mae(actual_dist,predicted_dist)

mer_list.append(mer_error)
mean_error_list.append(mean_error)
mse_list.append(mse_error)
mae_list.append(mae_error)

#print(mer_list)
#print(mean_error_list)
#print(mse_list)
#print(mae_list)
```
